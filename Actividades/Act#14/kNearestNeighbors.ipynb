```python
#Librerias a utilizar

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Cargar el dataset
dataframe = pd.read_csv(r"reviews_sentiment.csv",sep=';')
dataframe.head()
```

Preprocesamiento

```python
#Crear los datasets de entrenamiento y testeo

X = dataframe[['wordcount','sentimentValue']].values
y = dataframe['Star Rating'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0) #Dividir los datasets en entrenamiento y testeo
scaler = MinMaxScaler() #Rescala los valores de las features entre 0 y 1
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Crear el modelo

```python
#Buscar por el valor de k para los mejores resultados

k_range = range(1, 20)
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    scores.append(knn.score(X_test, y_test))

kmax = np.argmax(scores)+1
print(f'Mejor k = {kmax}, con score de: {round(scores[kmax-1]*100, 2)}%')
```

```python
#Definir el knn con mejor valor de k

n_neighbors = kmax

knn = KNeighborsClassifier(n_neighbors)
knn.fit(X_train, y_train)
print('Precisión de K-NN en el dataset de entrenamieto: {:.2f}%'
     .format(knn.score(X_train, y_train)*100))
print('Precisión de K-NN en el dataset de testeo: {:.2f}%'
     .format(knn.score(X_test, y_test)*100))
```
